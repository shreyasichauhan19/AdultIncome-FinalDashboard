```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("adult.csv") # this a data file that gets loaded

columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain",
           "capital-loss", "hours-per-week", "native-country", "annualsalary"]
df = pd.DataFrame(data = np.array(data), columns=columns)

df_less50k = df[df["annualsalary"] == " <=50K"]
temp = df_less50k.groupby("education")["education"].count()

pie, ax = plt.subplots(figsize=[10,6])
labels = temp.keys()
plt.pie(x=temp, autopct="%.1f%%", explode=[0.05]*len(temp), labels=labels, pctdistance=0.5)
plt.title("Education level distribution for people who makes <=50k per year")



```

```python
df_greater50k = df[df["annualsalary"] == " >50K"]
temp = df_greater50k.groupby("education")["education"].count()

pie, ax = plt.subplots(figsize=[10,6])
labels = temp.keys()
plt.pie(x=temp, autopct="%.1f%%", explode=[0.05]*len(temp), labels=labels, pctdistance=0.5)
plt.title("Education level distribution for people who makes >50k per year")
```
