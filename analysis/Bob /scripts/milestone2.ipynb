```python
def load_and_process(url_or_path_to_csv_file):
    df1 = (
        pd.read_csv(url_or_path_to_csv_file, usecols=['education-num','annualsalary'])
        .sort_values("education-num", ascending=True)
    )
    
    for i in range(16):
        df_e = df1.loc[lambda x: x['education-num']== i]
        print(i,df_e['annualsalary'].value_counts())
    
    df2 = (
        pd.read_csv(url_or_path_to_csv_file, usecols=['hours-per-week','annualsalary'])
        .sort_values("hours-per-week", ascending=True)
    )
    for i in range(100):
        df_h = df2.loc[lambda x: x['hours-per-week']== i]
        print(i,df_h['annualsalary'].value_counts())
    
    return df1, df2
    
```

```python
load_and_process("adult.csv")
```
